```python
dummy = ['Has Table booking','Has Online delivery']
rest1 = pd.get_dummies(rest1,columns=dummy,drop_first=True)
rest1.head()
```

```python
#Ratio between restaurants that allow table booking vs. those that do not allow table booking
```

```python
table_booking=rest1[rest1["Has Table booking_Yes"]==1]["Restaurant ID"].count()
table_notbooking=rest1[rest1["Has Table booking_Yes"]==0]["Restaurant ID"].count()
print("Ratio b/w has table booking vs non table bookin",round((table_booking/table_notbooking),2))
```

```python
print(table_booking,table_notbooking)
```

```python
import matplotlib.pyplot as plt
```

```python
labels = 'Table Booking', 'Table Not Booking'
sizes = [table_booking,table_notbooking]
explode = (0.1, 0)  

fig1, ax1 = plt.subplots(figsize=(9,9))
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',shadow=True, startangle=90)
ax1.set_title("Table Booking vs No Table Booking")
ax1.axis('equal')

plt.show()
```

```python
#Percentage of restaurant that has online delivery
```

```python
onlinedelivery_yes=rest1[rest1["Has Online delivery_Yes"]==1]["Restaurant ID"].count()
onlinedelivery_no=rest1[rest1["Has Online delivery_Yes"]==0]["Restaurant ID"].count()
print("Percentage of restaurants providing onlinedelivery :{}%".format(round((onlinedelivery_yes/len(rest1)),3)*100))
```

```python
labels = 'Online Delivery','No Online Delivery'
size = [onlinedelivery_yes,onlinedelivery_no]
explode = (0.1,0)
fig1,ax1 = plt.subplots(figsize=(9,9))
ax1.pie(size,explode=explode,labels=labels,autopct='%1.1f%%',shadow=True,startangle=90)
ax1.set_title("Online Delivery vs No Online Delivery")
ax1.axis('equal')
plt.show()
```

```python
rest.columns
cuisines = rest['Cuisines'].apply(lambda x: pd.Series(x.split(',')))
```

```python
cuisines.columns = ['Cuisine_1','Cuisine_2','Cuisine_3','Cuisine_4','Cuisine_5','Cuisine_6','Cuisine_7','Cuisine_8']
cuisines.head()
```

```python
cuisines = pd.concat([rest,cuisines],axis=1)
cuisines.head()
```

```python
cuisine_loc = pd.DataFrame(cuisines[['Country','City','Locality Verbose','Cuisine_1','Cuisine_2','Cuisine_3',
                                        'Cuisine_4','Cuisine_5','Cuisine_6','Cuisine_7','Cuisine_8']])
```

```python
cuisine_loc_stack=pd.DataFrame(cuisine_loc.stack())
cuisine_loc.head()
```

```python
keys = [c for c in cuisine_loc  if c.startswith('Cuisine')]
a=pd.melt(cuisine_loc, id_vars='Locality Verbose', value_vars=keys, value_name='Cuisines') 
#melting the stack into one row
max_rate=pd.DataFrame(a.groupby(by=['Locality Verbose','variable','Cuisines']).size().reset_index())
#find the highest restuarant in the city
max_rate
del max_rate['variable']
max_rate.columns=['Locality Verbose','Cuisines','Count']
max_rate.head()
```

```python
loc=max_rate.sort_values('Count', ascending=False).groupby(by=['Locality Verbose'],as_index=False).first()
loc.head()
```

```python
rating_res=loc.merge(rest,left_on='Locality Verbose',right_on='Locality Verbose',how='inner') 
```

```python
df=pd.DataFrame(rating_res[['Country','City','Locality Verbose','Cuisines_x','Count']]) 
```

```python
country=rating_res.sort_values('Count', ascending=False).groupby(by=['Country'],as_index=False).first()
```

```python
con=pd.DataFrame(country[['Country','City','Locality','Cuisines_x','Count']])
con.columns=['Country','City','Locality','Cuisines','Number of restaurants in the country']
```

```python
rest_cuisine = pd.DataFrame(cuisines[['Restaurant Name','City','Cuisine_1','Cuisine_2','Cuisine_3','Cuisine_4',
                                         'Cuisine_5','Cuisine_6','Cuisine_7','Cuisine_8']])
rest_cuisine_stack=pd.DataFrame(rest_cuisine.stack()) 
rest_cuisine.head()
```
